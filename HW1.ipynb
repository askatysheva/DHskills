```python
#Третье задание

thirdtask = "I can hack the FBI site"
letters = list(thirdtask)
print(letters)

#Третий символ строки
print(letters[2])

#Предпоследний список строки
print(letters[-2])

#Первые пять символов строки
print(letters[:5])

#Вся строка, кроме последних двух символов
print(letters[:-2])

#Все символы с четными индексами
print(letters[::2])

#Все символы с нечетными индексами
print(letters[1::2])

#Все символы в обратном порядке
print(letters[::-1])

#Все символы строки через один в обратном порядке, начиная с последнего
print(letters[::-2])

#Длина строки
print(len(letters))
```

```python
#Выражение, которое просит пользователя ввести с клавиатуры какое-нибудь слово
pomogite = input("Введите любле слово\n")

#Проверка наличия слова в массиве
if pomogite not in ruinit:
    ruinit.append(pomogite)
    print("Слово", pomogite, "добавлено")
    
else:
    thereis = ruinit.count(pomogite)
    print("Слово \"", pomogite, "\" встречается", thereis, "раз(-а)")
```
