## Задание №1
Необходимо выделить все биграммы в этом тексте и сохранить их список в какую-либо переменную.

```python
#Дан текст:
from string import punctuation 

text = """Блестящие колготки туго и соблазнительно облегали прекрасные бедра: чудесное дополнение к легкому вечернему платью. 
От самых кончиков бриллиантовых сережек до носков изящных туфелек на тонких шпильках: все было просто шикарно. 
Глаза с только что наведенными тенями рассматривали отражение в зеркале, и накрашенные яркой красной помадой губы растягивались 
от удовольствия. Внезапно сзади послышался детский голос: Папа?"""

from string import punctuation 
words = text.split() #Разбиваю текст на слова
words = [w.strip(punctuation) for w in words] #Удаляю пунктуацию из текста

bigrams = []
for l in range(len(words)-1): #Весь текст -1 элемент
    bigrams.append([words[l],words[l+1]]) #Разбиваем на биграммы
print(bigrams)
```
