### Задание №1 ###
Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
#Скачан текст «Литературных анекдотов»

import os, re
from pymystem3 import Mystem

def just_do_it(raz, dva):                          #Создаю функцию
    with open(raz, "r", encoding="utf-8") as a:
        smeshno = a.read()                         #Читаем файлы
   
    from string import punctuation 
    words = smeshno.split()                        #Разбиваю текст на слова
    words = [w.strip(punctuation) for w in words]  #Удаляю пунктуацию из текста
    antiprobel = [k for k in words if len(k) != 0] #Удаляю пустые эдементы
    obratno = " ".join(antiprobel)                 #Склеиваю в строку

    m = Mystem()                                   #Лемматизирую
    lemmas = m.lemmatize(obratno)
    lemlem = ''.join(lemmas)                       #Снова строка


    with open(dva, "a+", encoding="utf-8") as c:   #Создаю новый файл и открываю его для добавления и чтения
        c.write(lemlem + "\n")                     #Записываю то, что получилось, в новый файл

just_do_it(input("Путь к исходному файлу:\n"), input("Имя нового файла:\n"))    #Вызов функции
```
